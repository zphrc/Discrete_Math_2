Task 1:
Consider the following experiment:

1. The experiment consists of $n$ identical trials.
2. The outcome of each trial falls into one of $k$ classes or cells.
3. The probability that the outcome of a single trial falls into cell $i$, is $p_i,i = 1,2,...,k$ and remains the same from trial to trial. Notice that $p_1+p_2+p_3+···+p_k =1.$
4. The trials are independent.
5. The random variables of interest are $Y_1 , Y_2 , . . . , Y_k$, where $Y_i$ equals the number of trials for which the outcome falls into cell $i$. Notice that $Y_1 +Y_2 +Y_3 +···+Y_k =n$.
6. The joint probability distribution $P(y) = \frac{n!}{y_1!y_2!y_3!...y_n!} P_1^{y_1} P_2^{y_2} P_3^{y_3}....P_n^{y_n}$

Question 1: Show that the expected value $E[y] = \Sigma y \cdotp P(y) = n \cdot p_i$<br><br>

$\textbf{Solution:}$

The expected value $E[Y_i]$ of the random variable $Y_i$ is given by:

\begin{align*}
E[Y_i] &= \sum_{y=0}^{n} y \cdot P(Y_i = y) \\
       &= \sum_{y=0}^{n} y \cdot \frac{n!}{y!(n-y)!} p_i^y (1-p_i)^{n-y} \\
       &= \sum_{y=0}^{n} \frac{n!}{(y-1)!(n-y)!} p_i^y (1-p_i)^{n-y} \\
       &= \sum_{y=1}^{n} \frac{n!}{(y-1)!(n-y)!} p_i^y (1-p_i)^{n-y} \\
       &= \sum_{y=1}^{n} \frac{n!}{(y-1)!(n-y)!} p_i^{y-1}p_i (1-p_i)^{n-y} \\
       &= p_i \sum_{y=1}^{n} \frac{n!}{(y-1)!(n-y)!} p_i^{y-1} (1-p_i)^{n-y} \\
       &= p_i \sum_{y=0}^{n-1} \frac{n!}{y!(n-1-y)!} p_i^y (1-p_i)^{n-1-y} \\
       &= p_i \cdot 1 \\
       &= p_i \cdot n
\end{align*}

Therefore, the expected value $E[Y_i] = n \cdot p_i$.

If 500 adults are sampled  randomly, find the probability rhat the sample contains 100 person between 18 and 24, 200 between, 200 between the ages of 25 and 34, and 200 between the ages of 45 and 64.. What is the expected value for to obtain a person in the 65 and above?<br><br>

$\textbf{Solution:}$

The probability of obtaining 100 persons between 18 and 24, 200 between 25 and 34, and 200 between 45 and 64 can be calculated using the given proportions and the binomial distribution formula. \begin{align*}

Task 2: Consider the following experiment:

1. The experiment consists of a fixed number, $n$, of identical trials.
2. Each trial results in one of two outcomes: success, $S$, or failure, $F$.
3. The probability of success on a single trial is equal to some value $p$ and remains the same from trial to trial. The probability of a failure is equal to q = $(1 − p)$.
4. The trials are independent.
5. The random variable of interest is $Y$ , the number of successes observed during the $n$ trials.

Question 1: From the following steps shown above, derive the probability distribution of the experiment.<br><br>

$\textbf{Solution:}$

The probability of observing $k$ successes in $n$ trials is given by the binomial distribution:

\begin{align*}
P(Y = k) = \binom{n}{k} p^k q^{n-k}
\end{align*}

where $\binom{n}{k}$ is the binomial coefficient, $p$ is the probability of success, and $q = (1 - p)$ is the probability of failure.

The expectation of the binomial distribution is given by:

\begin{align*}
E[Y] = \sum_{k=0}^{n} k \cdot P(Y = k) = \sum_{k=0}^{n} k \cdot \binom{n}{k} p^k q^{n-k}
\end{align*}

Using the properties of binomial coefficients, we find that:

\begin{align*}
E[Y] = np
\end{align*}

What is the probability that at least nine of ten receiving the medication will recover?<br><br>

We are given $p = 0.3$ and $n = 10$. Using the binomial distribution formula, we calculate the probability of observing at least nine successes:

\begin{align*}
P(Y \geq 9) = \sum_{k=9}^{10} \binom{10}{k} (0.3)^k (0.7)^{10-k}
\end{align*}

Task 3: Consider a probility distribution of $p(y) = \frac{\lambda^y}{y!} e^{-\lambda}$.

Question 1: Find a general formula for the expected value of this distribution.
<br><br>

$\textbf{Solution:}$

Consider a probability distribution given by $P(y) = \frac{\lambda^y}{y!} e^{-\lambda}$.

To find the expected value of this distribution, we use the definition of expected value:

\begin{align*}
E[Y] &= \sum_{y=0}^{\infty} y \cdot P(y) \\
     &= \sum_{y=0}^{\infty} y \cdot \frac{\lambda^y}{y!} e^{-\lambda} \\
     &= e^{-\lambda} \sum_{y=0}^{\infty} \frac{\lambda^y}{(y-1)!} \\
     &= e^{-\lambda} \lambda \sum_{y=1}^{\infty} \frac{\lambda^{y-1}}{(y-1)!} \\
     &= e^{-\lambda} \lambda \sum_{k=0}^{\infty} \frac{\lambda^k}{k!} \\
     &= e^{-\lambda} \lambda e^{\lambda} \\
     &= \lambda
\end{align*}

Therefore, the general formula for the expected value of this distribution is $E[Y] = \lambda$.

Question 2: A certain type of tree has seedlings randomly dispersed in a large area, with the mean density of seedlings being approximately five per square yard. If a forester randomly locates ten 1-square-yard sampling regions in the area, find the probability that none of the regions will contain seedlings.<br><br>

$\textbf{Solution:}$

Given that the mean density of seedlings is approximately five per square yard, we have $\lambda = 5$. Let $X$ be the number of sampling regions without seedlings.

The probability that none of the regions will contain seedlings is given by:

\begin{align*}
P(X = 10) &= \frac{\lambda^0}{0!} e^{-\lambda} \\
          &= e^{-5}
\end{align*}

Therefore, the probability that none of the regions will contain seedlings is $e^{-5}$.
