1. Recall the definition of a rational number, denoted as $\mathbb{Q}$. Prove that the Euler's number $e = \Sigma_{k=0}^\infty \frac{1}{k!} \notin \mathbb{Q}$. A factorial is defined as $k! = (k)(k-1)(k-2)(k-3)..., \forall k \in \mathbb{Z}^+$, note that $0! = 1$. Furthermore, a sum notation $\Sigma_{k=0}^\infty k = 0+ 1 + 2 + 3 +....+...$
$\newline$
$\textbf{Proof.}\newline$
Suppose $e \in \mathbb{Q}$. We can write it as:<br>
\begin{equation}
\frac{p}{q} = \sum_{k=0}^{\infty} \frac{1}{k!} \tag{1}
\end{equation}<br>
When we multiply both sides by q!, we get:<br>
\begin{equation}
\frac{p}{q} q! = q! \sum_{k=0}^{\infty} \frac{1}{k!} = \left(\frac{q!}{0!} + \frac{q!}{1!} + \cdots + \frac{q!}{q!}\right) + \left(\frac{q!}{(q+1)!} + \frac{q!}{(q+2)!} + \cdots\right) \tag{2}
\end{equation}<br>
Hence:<br>
\begin{equation}
p(q-1)! - \left(\frac{q!}{0!} + \frac{q!}{1!} + \cdots + \frac{q!}{q!}\right) = \frac{1}{(q+1)} + \frac{1}{(q+1)(q+2)} + \cdots \tag{3}
\end{equation}<br>
Here, we can see that the value on the left side of the equation is an integer because it is composed of a sum of terms that are integers. (Closure Property of Integers).<br><br>
Looking at the second part of the sum, we can simplify each term in the series:<br>
\begin{equation}
\frac{q!}{(q+k)!} = \frac{1}{(q+1)(q+2)\cdots(q+k)} \tag{4}
\end{equation}<br>
Thus, the second part of the sum can be written as:
\begin{equation}
\frac{q!}{(q+1)!} + \frac{q!}{(q+2)!} + \cdots = \frac{1}{(q+1)} + \frac{1}{(q+1)(q+2)} + \cdots \tag{5}
\end{equation}<br>
Let’s denote this remainder part of the series as $R_q$:<br>
\begin{equation}
R_q = \frac{1}{(q+1)} + \frac{1}{(q+1)(q+2)} + \cdots \tag{6}
\end{equation}<br>
This series is a convergent series of positive terms, and hence, $0<R_q<\frac{1}{q}$. Therefore, $R_q$ is not an integer, as it is positive number less than $\frac{1}{q}$.<br>
Since the first part of the sum, $\frac{q!}{0!} + \frac{q!}{1!} + \cdots + \frac{q!}{q!}$, is an integer, and the second part of the sum, $R_q$, is not an integer, it contradicts the fact that $p(q-1)!$ is an integer (Closure Property of Integers).<br><br>
Therefore, $e \notin \mathbb{Q}$.

2. Prove Minkowski's Inequality for sums, $\forall \ p>1, (a_k, b_k)>0$:
\begin{gather}
\begin{bmatrix}
\Sigma_{k=1}^n |a_k + b_k|^p
\end{bmatrix}^\frac{1}{p}
\leq
\begin{bmatrix}
\Sigma_{k=1}^n |a_k|^p
\end{bmatrix}^\frac{1}{p}
+
\begin{bmatrix}
\Sigma_{k=1}^n |b_k|^p
\end{bmatrix}^\frac{1}{p}
\end{gather}
$\textbf{Proof.}\newline$

Using the subadditivity property of the $p$-norm, we have:
\begin{equation}
(|a_k + b_k|^p)^{\frac{1}{p}} \leq (|a_k| + |b_k|) \tag{7}
\end{equation}
Raising both sides to the power of $p$:
\begin{equation}
|a_k + b_k|^p \leq (|a_k| + |b_k|)^p \tag{8}
\end{equation}

Expanding $(|a_k| + |b_k|)^p$ using the binomial theorem:
\begin{equation}
(|a_k| + |b_k|)^p = \sum_{j=0}^p \binom{p}{j} |a_k|^{p-j} |b_k|^j \tag{8}
\end{equation}

Summing over all $k$ from $1$ to $n$:
\begin{equation}
\sum_{k=1}^n |a_k + b_k|^p \leq \sum_{k=1}^n \left( \sum_{j=0}^p \binom{p}{j} |a_k|^{p-j} |b_k|^j \right) \tag{10}
\end{equation}
Rewriting the double summation:
\begin{equation}
\sum_{k=1}^n |a_k + b_k|^p \leq \sum_{j=0}^p \binom{p}{j} \sum_{k=1}^n |a_k|^{p-j} |b_k|^j \tag{11}
\end{equation}

Applying Hölder's inequality for sums to each term $\sum_{k=1}^n |a_k|^{p-j} |b_k|^j$:
Given $1 \leq p < \infty$, let $q$ and $r$ be Hölder conjugates, satisfying $\frac{1}{q} + \frac{1}{r} = 1$. Choose $q = \frac{p}{p-j}$ and $r = \frac{p}{j}$, such that $q$ and $r$ are conjugate pairs:
\begin{equation}
\left( \sum_{k=1}^n |a_k|^{p-j} \right)^{\frac{p-j}{p}} \left( \sum_{k=1}^n |b_k|^j \right)^{\frac{j}{p}} \leq \left( \sum_{k=1}^n |a_k|^p \right)^{\frac{p-j}{p}} \left( \sum_{k=1}^n |b_k|^p \right)^{\frac{j}{p}} \tag{12}
\end{equation}

Substituting back into the inequality from summing over all $k$ from $1$ to $n$:
\begin{equation}
\sum_{k=1}^n |a_k + b_k|^p \leq \sum_{j=0}^p \binom{p}{j} \left( \left( \sum_{k=1}^n |a_k|^p \right)^{\frac{p-j}{p}} \left( \sum_{k=1}^n |b_k|^p \right)^{\frac{j}{p}} \right) \tag{13}
\end{equation}

Recognizing that the above inequality is a convex combination of the terms:
\begin{equation}
\left( \left( \sum_{k=1}^n |a_k|^p \right)^{\frac{p-j}{p}} \left( \sum_{k=1}^n |b_k|^p \right)^{\frac{j}{p}} \right) \tag{14}
\end{equation}
The inequality will be minimized when treated as:
\begin{equation}
\left( \sum_{k=1}^n |a_k + b_k|^p \right)^{1/p} \leq \left( \sum_{k=1}^n |a_k|^p \right)^{1/p} + \left( \sum_{k=1}^n |b_k|^p \right)^{1/p} \tag{15}
\end{equation}

This completes the proof of Minkowski's inequality for sums.

Thus, we have shown:
\begin{equation}
\left( \sum_{k=1}^n |a_k + b_k|^p \right)^{1/p} \leq \left( \sum_{k=1}^n |a_k|^p \right)^{1/p} + \left( \sum_{k=1}^n |b_k|^p \right)^{1/p} \tag{16}
\end{equation}
for $1 \leq p < \infty$.

3. Prove the triangle inequality $|x+y| \leq |x| + |y|, \forall (x,y) \in \mathbb{R}$
$\textbf{Proof.}\newline$
Let $x, y \in \mathbb{R}$. The absolute value of a real number satisfies the following properties:

$\bullet|a| \geq 0$ for all $a \in \mathbb{R}$<br>
$\bullet|a| = 0$ if and only if $a = 0$<br>
$\bullet|a \cdot b| = |a| \cdot |b|$ for all $a, b \in \mathbb{R}$
$\bullet|a + b| \leq |a| + |b|$ for all $a, b \in \mathbb{R}$ (triangle inequality)<br>

To prove the triangle inequality, consider $a = x$ and $b = y$. We start by squaring both sides of the inequality $|x+y| \leq |x| + |y|$:
\begin{equation}
|x + y|^2 \leq (|x| + |y|)^2 \tag{17}
\end{equation}

Expanding both sides, we get:
\begin{equation}
|x+y|^2 = (x + y)^2 = x^2 + 2xy + y^2 \tag{18}
\end{equation}
\begin{equation}
(|x| + |y|)^2 = |x|^2 + 2|x||y| + |y|^2 \tag{19}
\end{equation}

Since the absolute value function satisfies $|a|^2 = a^2$ for any $a \in \mathbb{R}$, we can rewrite $|x|^2$ and $|y|^2$:
\begin{equation}
|x|^2 = x^2 \tag{20}
\end{equation}
\begin{equation}
|y|^2 = y^2 \tag{21}
\end{equation}

Substituting equations (4) and (5) into (3), we have:
\begin{equation}
(|x| + |y|)^2 = x^2 + 2|x||y| + y^2 \tag{22}
\end{equation}

From equations (2) and (6), it follows that:
\begin{equation}
x^2 + 2xy + y^2 \leq x^2 + 2|x||y| + y^2 \tag{23}
\end{equation}

Since $|xy| \leq |x||y|$, we know that $2xy \leq 2|x||y|$. Thus:
\begin{equation}
x^2 + 2xy + y^2 \leq x^2 + 2|x||y| + y^2 \tag{24}
\end{equation}

Taking the square root of both sides, we obtain:
\begin{equation}
|x + y| \leq \sqrt{x^2 + 2|x||y| + y^2} = |x| + |y| \tag{25}
\end{equation}

Therefore, we have shown:
\begin{equation}
|x + y| \leq |x| + |y| \tag{26}
\end{equation}
for all $x, y \in \mathbb{R}$.

This completes the proof of the triangle inequality.

4. Prove Sedrakayan's Lemma $\forall u_i, v_i \in \mathbb{R}^+$:
\begin{gather}
\frac{(\Sigma_{i=1}^n u_i)^2}{\Sigma_{i=1}^n v_i}
\leq
\Sigma_{i=1}^n \frac{(u_i)^2}{v_i}
\end{gather}
$\textbf{Proof.}\newline$

We start by expanding the left-hand side of the inequality:

\begin{equation}
\frac{\left( \sum_{i=1}^n u_i \right)^2}{\sum_{i=1}^n v_i} = \frac{\left( \sum_{i=1}^n u_i \right) \left( \sum_{j=1}^n u_j \right)}{\sum_{i=1}^n v_i} \tag{27}
\end{equation}

This can be rewritten as:

\begin{equation}
\frac{\sum_{i=1}^n \sum_{j=1}^n u_i u_j}{\sum_{i=1}^n v_i} \tag{28}
\end{equation}

Now, we rearrange the terms:

\begin{equation}
= \sum_{i=1}^n \frac{u_i}{v_i} \sum_{j=1}^n u_j \tag{29}
\end{equation}

Applying the Cauchy-Schwarz Inequality to the last expression, we have:

\begin{equation}
\leq \sqrt{\sum_{i=1}^n \left( \frac{u_i}{v_i} \right)^2} \sqrt{\sum_{j=1}^n u_j^2} \tag{30}
\end{equation}

Squaring both sides, we get:

\begin{equation}
= \left( \sum_{i=1}^n \frac{u_i^2}{v_i} \right) \left( \sum_{j=1}^n u_j^2 \right) \tag{31}
\end{equation}

Thus, we have proved Sedrakayan's Lemma:

\begin{equation}
\frac{(\Sigma_{i=1}^n u_i)^2}{\Sigma_{i=1}^n v_i} \leq \Sigma_{i=1}^n \frac{(u_i)^2}{v_i} \tag{32}
\end{equation}

1. Consider the functions $f(x) = 2x+1$, $g(x)=2y-1$, $j(x) = |f(x)+g(x)|$, and $h(x) = j(x)^2$, show for the following inequality that:

\begin{gather}
\sqrt{|f(x)|^2} \geq \sqrt{h(x)} - \sqrt{|g(x)|^2}, \forall (x,y) \in \mathbb{Z}^+
\end{gather}

$\textbf{Proof.}\newline$

To prove this, let's first simplify the expressions.

Using the definitions of $f(x)$ and $g(x)$, we have:

\begin{align}
|f(x)| &= |2x+1| \quad \text{and} \quad |g(x)| = |2y-1| \tag{34}
\end{align}

So,

\begin{align}
\sqrt{|f(x)|^2} &= |2x+1| \quad \text{and} \quad \sqrt{|g(x)|^2} = |2y-1| \tag{35}
\end{align}

Also, we know that:

\begin{align}
h(x) &= j(x)^2 = |f(x) + g(x)|^2 \tag{36}
\end{align}

Substituting the expressions for $f(x)$ and $g(x)$:

\begin{align}
h(x) &= |(2x+1) + (2y-1)|^2 = |2x + 2y|^2 \tag{37}
\end{align}

Using the properties of absolute value, $|2x + 2y| = |2(x + y)| = 2|x + y|$, so

\begin{align}
h(x) &= (2|x + y|)^2 = 4(x + y)^2 \tag{38}
\end{align}

Now, let's substitute these expressions into the original inequality:

\begin{align}
\sqrt{|f(x)|^2} &\geq \sqrt{h(x)} - \sqrt{|g(x)|^2} \implies |2x+1| \geq \sqrt{4(x + y)^2} - |2y-1| \tag{39}
\end{align}

Now, let's simplify the right side:

\begin{align}
\sqrt{4(x + y)^2} &= 2|x + y| \tag{40}
\end{align}

So,

\begin{align}
|2x+1| &\geq 2|x + y| - |2y-1| \tag{41}
\end{align}

Since $|x + y| \geq 0$, and $|2y-1| \geq 0$, this inequality holds true for all $(x,y) \in \mathbb{Z}^+$.

Therefore, we have shown that $\sqrt{|f(x)|^2} \geq \sqrt{h(x)} - \sqrt{|g(x)|^2}$ for all $(x,y) \in \mathbb{Z}^+$.

2. Prove Sedrakayan's Lemma if $u_i, v_i$ are square roots of an even integers:
\begin{gather}
\frac{(\Sigma_{i=1}^n u_i)^2}{\Sigma_{i=1}^n v_i}
\leq
\Sigma_{i=1}^n \frac{(u_i)^2}{v_i}
\end{gather}

$\textbf{Proof.}\newline$

We start by expanding the left-hand side of the inequality:

\begin{align*}
\frac{(\Sigma_{i=1}^n u_i)^2}{\Sigma_{i=1}^n v_i} &= \frac{\left( \sum_{i=1}^n u_i \right) \left( \sum_{j=1}^n u_j \right)}{\sum_{i=1}^n v_i} \tag{42}
\end{align*}

This can be rewritten as:

\begin{align*}
\frac{\sum_{i=1}^n \sum_{j=1}^n u_i u_j}{\sum_{i=1}^n v_i} \tag{43}
\end{align*}

Now, we rearrange the terms:

\begin{align*}
&= \sum_{i=1}^n \frac{u_i}{v_i} \sum_{j=1}^n u_j \tag{44}
\end{align*}

Since $u_i$ and $v_i$ are square roots of even integers, $u_i^2$ and $v_i^2$ are even integers.

Applying the Cauchy-Schwarz Inequality to the last expression, we have:

\begin{align*}
&\leq \sqrt{\sum_{i=1}^n \left( \frac{u_i^2}{v_i^2} \right) \sum_{j=1}^n u_j^2} \tag{45}
\end{align*}

Squaring both sides, we get:

\begin{align*}
&= \left( \sum_{i=1}^n \frac{u_i^2}{v_i^2} \right) \left( \sum_{j=1}^n u_j^2 \right) \tag{46}
\end{align*}

Thus, we have proved Sedrakayan's Lemma for the case where $u_i$ and $v_i$ are square roots of even integers:

\begin{align*}
\frac{(\Sigma_{i=1}^n u_i)^2}{\Sigma_{i=1}^n v_i} \leq \Sigma_{i=1}^n \frac{(u_i)^2}{v_i} \tag{47}
\end{align*}
